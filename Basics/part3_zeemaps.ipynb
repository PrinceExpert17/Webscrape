# Map ID
- We are going to create a variable to store the mapdata function
- This variable will loop through all id's and get the first 20

```python
# Scrape
map_data = getmapdata()

add_id = []
for item in map_data[0:20]:
    _id = item['id']
    print(_id)
    idlist = getadddetail(_id)
    add_id.append(idlist)
```

# Extraction
- We are then going to get the information from the data and store it in a list

```python
htmldetail = []
for ad in add_id:
    datadict = {}
    html = ad['t']
    soup = bs(html, 'html.parser')
    try:
        datadict['url'] = soup.find('a')['href']
    except:
        datadict['url'] = ''
    try:
        datadict['number'] = soup.find('span', {'class':'phone'}).text
    except:
        datadict['number'] = ''
    for item in soup.find_all('span'):
        datadict[item.text] = str(item.next_sibling).replace('xa0', '').strip()
    htmldetail.append(datadict)
htmldetail
```

```python
master = []
for ad, html in zip(add_id, htmldetail):
    output = {**ad,**html}
    output = {**output, **ad['ad']}
    del output['ad'] 
    master.append(output)
master
```

# Conversion to csv 
- Convert to a dataframe and store as a csv file

```python
# Convert to dataframe
panda = pd.DataFrame(master)
#panda.to_csv('./files/Housing_Map.csv')
panda.head(10)
```
